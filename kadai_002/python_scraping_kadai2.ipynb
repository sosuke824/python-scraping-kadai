```python
import requests
import pprint
import getpass

# 1. APIキーの入力（セキュリティのため伏せ字で入力）
api_key = getpass.getpass("Google Maps APIキーを入力してください: ")

# 2. 駅名の入力
station_name = input("検索したい駅名を入力してください（例: 東京駅）: ")

# --- ステップ1: 駅名から緯度・経度を取得 (Geocoding API) ---
geo_url = "https://maps.googleapis.com/maps/api/geocode/json"
geo_params = {
    "address": station_name,
    "key": api_key,
    "language": "ja"
    }

geo_response = requests.get(geo_url, params=geo_params).json()

# pprint.pprint(geo_response)

location = geo_response["results"][0]["geometry"]["location"]
lat_lng = f"{location['lat']},{location['lng']}"

# --- ステップ2: 半径500m以内のレストランを検索 (Places API - Nearby Search) ---
places_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
places_params = {
    "location": lat_lng,
    "radius": 500,
    "type": "restaurant",
    "key": api_key,
    "language": "ja"
    }

places_response = requests.get(places_url, params=places_params).json()

# pprint.pprint(places_response)

if places_response["status"] == "ZERO_RESULTS":
  print("指定された範囲内にレストランが見つかりませんでした。")
elif places_response["status"] != "OK":
  print(f"エラー: レストラン情報の取得に失敗しました。({places_response['status']})")
else:
  for place in places_response.get("results", []):
    name = place.get("name")
    rating = place.get("rating", "評価なし")
    address = place.get("vicinity")
    print(name, rating, address)
```
